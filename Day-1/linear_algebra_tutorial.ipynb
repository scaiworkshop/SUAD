## Vector Magnitude and Unit Vectors


````python
# Create the vector as NumPy array
u = np.array([1,3])

# Find the norm using np.linalg.norm() function
norm = np.linalg.norm(u)
print(norm**2)

# Find unit vector
u_hat= u / np.linalg.norm(u)
print(np.linalg.norm(u_hat)**2)
````

### Vector Distance

Distance is a relational concept. It refers to the length (or norm) of the difference between two vectors. Hence, we use norms and lengths to measure the distance between vectors. Consider the vectors $x$
 and $y$
, we define the distance $d(x,y)$
 as:

 \begin{align}
 d(x,y) := \lVert x - y \rVert = \sqrt{\langle x - y, x - y \rangle}
 \end{align}

````python
x, y = np.array([-2,2]), np.array([4,-3])

print(x.shape)
print(y.shape)

np.dot(x,y)
````

````python
distance = np.linalg.norm(x-y, 2)
print(f'distance between x and y : {distance}')
````

## Vector angles and orthogonality

In machine learning, the angle between a pair of vectors is used as a measure of vector similarity.

Some math to understand what's going on 

The definition of the angle between vectors can be thought as a generalization of the law of cosines in trigonometry, which defines for a triangle with sides $a$
, $b$ , and $c$ , and an angle $\theta$ are related as:

$ \Vert x - y \Vert^2 = \Vert x \Vert^2 + \Vert y \Vert^2 - 2(\Vert x \Vert \Vert y \Vert) \cos \theta $

$\cos \theta = \frac{\langle x, y \rangle}{\Vert x \Vert \Vert y \Vert}$


image.png

````python
x, y = np.array([1, 2]), np.array([5, 7])

# here we translate the cos(theta) definition
cos_theta = (np.dot(x, y)) / (np.linalg.norm(x,2) * np.linalg.norm(y,2))
print(f'cos of the angle = {np.round(cos_theta, 3)}')
````

````python
cos_inverse = np.arccos(cos_theta)
print(f'angle in radiants = {np.round(cos_inverse, 3)}')
````

````python
degrees = cos_inverse * ((180)/np.pi)
print(f'angle in degrees = {np.round(degrees, 3)}')
````

### Systems of Linear Equations

The purpose of linear algebra as a tool is to solve systems of linear equations. Informally, this means to figure out the right combination of linear segments to obtain an outcome. Even more informally, think about making pancakes: In what proportion ($w_i \in \mathbb{R}$
) we have to mix ingredients to make pancakes? You can express this as a linear equation:

\
\begin{align}
f_\text{flour} \times w_1 + b_\text{baking powder}  \times w_2 + e_\text{eggs}  \times w_3 + m_\text{milk} \times w_4 = P_\text{pancakes}
\end{align}

\
The above expression describe a linear equation. A system of linear equations involve multiple equations that have to be solved simultaneously. Consider:


\begin{align}
x + 2y = 8 \\
5x - 3y = 1
\end{align}
\
Now we have a system with two unknowns, $x$ and $y$. Well see general methods to solve systems of linear equations later. For now, Ill give you the answer: $x=2$ and $y=3$. Geometrically, we can see that both equations produce a straight line in the 2-dimensional plane.

````python
df = pd.DataFrame({"x1": [0, 2], "y1":[8, 3], "x2": [0.5, 2], "y2": [0, 3]})
````

````python
plt.plot(df["x1"], df["y1"], label="1st Equation")
plt.plot(df["x2"], df["y2"], label="2nd Equation")

plt.axis('equal')
plt.plot([0, 2],[0, 0],'k-')
plt.plot([0, 0],[0, 8],'k-')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.title("System of Linear Equations Example")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````

Can we do the same with ```numpy```?

````python
a = np.array([[1, 2], [5, -3]])
b = np.array([8, 1])
solution = np.linalg.solve(a, b)
print("x = ", solution[0])
print("y = ", solution[1])
````

In the given code, matrix P has a shape of (3, 2), and matrix Q has a shape of (2, 4). To perform matrix multiplication Q multiplied by P, you need to transpose the matrices to make the dimensions compatible :

````python
result2 = np.dot(Q.T, P.T)  # Transpose Q and P for correct multiplication
print("Result of Q dot P:")
print(result2)
````
