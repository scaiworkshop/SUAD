## Image Filtering of different size

```python
#applying 3X3 mask
import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np

#creating the image
img = np.zeros([230, 233])
img = img.astype(np.uint8) 
for i in range(10, 220):
    j = 0
    while j <= 192:
        for k in range(j+0, j+16):
            img[i, k] = 0
        for k in range(j+17, j+24):
            img[i, k] = 255
        j += 24

m, n = img.shape 

img_new = np.zeros([m, n]) 


# Convolve the 3X3 mask over the image 
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = 0
        for l in range(-1, 2):
            for k in range(-1, 2):
                temp = temp + img[i+l, j+k]
        temp = temp/9
        
        img_new[i, j]= temp
    
img_new = img_new.astype(np.uint8) 


# Display the images
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) 
plt.show()

# Display the images
plt.imshow(cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
#applying 5X5 mask
import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np


m, n = img.shape 

img_new = np.zeros([m, n]) 
# Convolve the 5X5 mask over the image 
for i in range(3, m-3): 
    for j in range(3, n-3): 
        temp = 0
        for l in range(-2, 3):
            for k in range(-2, 3):
                temp = temp + img[i+l, j+k]
        temp = temp/25
        
        img_new[i, j]= temp 

img_new = img_new.astype(np.uint8) 


# Display the images
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) 
plt.show()

# Display the images
plt.imshow(cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
#applying 9X9 mask

import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np


m, n = img.shape 
img_new = np.zeros([m, n]) 
# Convolve the 9X9 mask over the image 
for i in range(4, m-4): 
    for j in range(4, n-4): 
        temp = 0
        for l in range(-4, 5):
            for k in range(-4, 5):
                temp = temp + img[i+l, j+k]
                
        temp = temp/81
                
        
        img_new[i, j]= temp 

img_new = img_new.astype(np.uint8) 


# Display the images
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) 
plt.show()

# Display the images
plt.imshow(cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
# Smoothing Uniform Filter
# Read the image 
img = cv2.imread('/content/board.png', 0) 

# Obtain number of rows and columns 
# of the image 
m, n = img.shape 

# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9

# Convolve the 3X3 mask over the image 
img_new = np.zeros([m, n]) 

for i in range(1, m-1): 
	for j in range(1, n-1): 
		temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
		
		img_new[i, j]= temp 
		
img_new = img_new.astype(np.uint8) 


# Display the images
plt.imshow(img) 
plt.show()

# Display the images
plt.imshow(img_new) 
plt.show()
```

## Edge Detection

```python
# Read The Image

imageMascot = cv2.imread("/content/mascot21.jfif") 
print(imageMascot.shape)
imageMascot_rgb = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2RGB)
plt.imshow(imageMascot_rgb)
plt.title('RGBMascot')
plt.figure()
```

```python
imageMascot_gray = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2GRAY)

plt.imshow(imageMascot_gray, cmap = plt.cm.gray)

plt.title('GrayMascot')
plt.figure()
```

```python
#to detect edge
# Canny Edge Detection

edges_img = cv2.Canny(imageMascot_gray, threshold1=100, threshold2=200) # Canny Edge Detection

# Display Canny Edge Detection Image
plt.imshow(edges_img, cmap = plt.cm.gray)

plt.title('Canny Edge Gray Mascot')
plt.figure()
```

```python

#to detect edge
# Sobel Edge Detection

Sobeledges_imgx = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=1, dy = 0, ksize=3) # sobel Edge Detection
plt.imshow(Sobeledges_imgx, cmap = plt.cm.gray)

plt.title('X Sobel EDGE GrayMascot')
plt.figure()

Sobeledges_imgy = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=0, dy = 1, ksize=3) # sobel Edge Detection

plt.imshow(Sobeledges_imgy, cmap = plt.cm.gray)

plt.title('Y Sobel EDGE GrayMascot')
plt.figure()


sobelxy = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3) # Combined X

# Display Canny Edge Detection Image
plt.imshow(sobelxy, cmap = plt.cm.gray)

plt.title('XY Sobel EDGE GrayMascot')
plt.figure()

```

## Image Compression

```python
# Read The Image

image = cv2.imread("/content/dog.jpg") 
print(image.shape)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Dog')
plt.figure()
```

```python
X = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(X, cmap = plt.cm.gray)

plt.title('Gray Dog')
plt.figure()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)

j = 0
for r in (5, 20, 50, 75, 100):
    # Construct approximate image
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```
